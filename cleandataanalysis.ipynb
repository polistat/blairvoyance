```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt

# plt.setstyle('seaborn')

df = pd.read_excel("./cleandataall.xlsx")

print('Columns: ' + str(df.columns.values))

plt.scatter(df['Partisanship'], df['DemMargin'])
plt.title('OG Partisanship vs DemMargin')
plt.show()

m, c, r, p, stder = stats.linregress(np.array(df['Partisanship']), np.array(df['DemMargin']))
resids = df['DemMargin'] - (m * df['Partisanship'] + c)
plt.scatter(df['Partisanship'], resids)
plt.title('Partisanship vs DemMargin Residuals')
plt.show()
```
